```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Neeraj\Downloads\Student_Performance.csv")
```

```python
print(df.columns)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df["Hours Studied"]=lb.fit_transform(df["Hours Studied"])
df["Previous Scores"]=lb.fit_transform(df["Previous Scores"])
df["Extracurricular Activities"] = lb.fit_transform(df["Extracurricular Activities"])
df["Sleep Hours	"]=lb.fit_transform(df["Sleep Hours"])
df["Sample Question Papers Practiced"]=lb.fit_transform(df["Sample Question Papers Practiced"])
```

```python
df
```

```python
y = df["Performance Index"]
X = df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(lr,X,y,cv=5)
```

```python
import pickle
with open('model.pickle','wb') as model_file:
    pickle.dump(lr,model_file)
```
